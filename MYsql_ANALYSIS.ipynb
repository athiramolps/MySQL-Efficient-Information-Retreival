# 1. List all movies with their details.


```python
SELECT * FROM `NetflixOriginals`
```

# 2. Find all movies released in 2020.


```python
SELECT * FROM `NetflixOriginals` WHERE Premiere LIKE '%2020%';
```

# 3. Get the average IMDB score of all movies.


```python
SELECT AVG(`IMDB Score`) AS Average_IMDB_Score FROM  `NetflixOriginals`;
```

# 4. Find the movie with the longest runtime.


```python
SELECT * FROM `NetflixOriginals` ORDER BY Runtime DESC LIMIT 1;
```

# 5. List all movies in the 'Comedy' genre.


```python
SELECT * FROM `NetflixOriginals` WHERE Genre LIKE '%Comedy%';
```

# 6. Count the number of movies in each genre.


```python
SELECT Genre, COUNT(*) AS Number_of_Movies FROM `NetflixOriginals` GROUP BY Genre;
```

# 7. Find the highest rated movie.


```python
SELECT * FROM `NetflixOriginals` ORDER BY `IMDB Score` DESC LIMIT 1;
```

# 8. List all movies not in English.


```python
SELECT * FROM `NetflixOriginals` WHERE Language NOT LIKE '%English%';
```

# 9. List movies with an IMDB score between 4.0 and 4.5.


```python
SELECT * FROM `NetflixOriginals` WHERE `IMDB Score` BETWEEN 4.0 AND 4.5;
```
